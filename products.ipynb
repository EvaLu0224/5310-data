```python
detailed_grocery_products = {
    "Beverages": [
        {"type": "Coffee", "min_price": 3, "max_price": 15},
        {"type": "Black Tea", "min_price": 3, "max_price": 7},
        {"type": "Green Tea", "min_price": 3, "max_price": 7},
        {"type": "Herbal Tea", "min_price": 3, "max_price": 8},
        {"type": "Cola", "min_price": 1, "max_price": 3},
        {"type": "Lemon-Lime Soda", "min_price": 1, "max_price": 3},
        {"type": "Root Beer", "min_price": 1, "max_price": 4},
        {"type": "Apple Juice", "min_price": 2, "max_price": 5},
        {"type": "Orange Juice", "min_price": 2, "max_price": 5},
        {"type": "Grape Juice", "min_price": 2, "max_price": 6},
        {"type": "Cranberry Juice", "min_price": 3, "max_price": 7},
        {"type": "Bottled Water", "min_price": 1, "max_price": 3},
        {"type": "Energy Drinks", "min_price": 2, "max_price": 5},
    ],
        
    "Snacks": [
        {"type": "Chips", "min_price": 2, "max_price": 5},
        {"type": "Nuts & Seeds", "min_price": 3, "max_price": 10},
        {"type": "Cookies", "min_price": 2, "max_price": 6},
        {"type": "Candy", "min_price": 1, "max_price": 5},
        {"type": "Crackers", "min_price": 2, "max_price": 5},
        {"type": "Popcorn", "min_price": 2, "max_price": 6}
    ],
    "Breakfast Foods": [
        {"type": "Cereal", "min_price": 3, "max_price": 8},
        {"type": "Oatmeal", "min_price": 2, "max_price": 5},
        {"type": "Pancake Mix", "min_price": 2, "max_price": 6},
        {"type": "Granola Bars", "min_price": 3, "max_price": 10},
        {"type": "Muffins", "min_price": 3, "max_price": 7},
        {"type": "Breakfast Sausage", "min_price": 3, "max_price": 8}
    ],
    "Canned & Packaged Foods": [
        {"type": "Canned Green Beans", "min_price": 1, "max_price": 3},
        {"type": "Canned Corn", "min_price": 1, "max_price": 3},
        {"type": "Canned Peas", "min_price": 1, "max_price": 3},
        {"type": "Canned Carrots", "min_price": 1, "max_price": 3},
        {"type": "Canned Spinach", "min_price": 2, "max_price": 4},
        {"type": "Canned Mixed Vegetables", "min_price": 2, "max_price": 4},
        {"type": "Canned Peaches", "min_price": 2, "max_price": 4},
        {"type": "Canned Pineapple", "min_price": 2, "max_price": 5},
        {"type": "Canned Pears", "min_price": 2, "max_price": 4},
        {"type": "Canned Mandarin Oranges", "min_price": 2, "max_price": 4},
        {"type": "Canned Fruit Cocktail", "min_price": 2, "max_price": 4},
        {"type": "Canned Cherries", "min_price": 3, "max_price": 5},
        {"type": "Canned Soup", "min_price": 2, "max_price": 5},
        {"type": "Pasta", "min_price": 1, "max_price": 3},
        {"type": "Rice", "min_price": 2, "max_price": 6},
        {"type": "Instant Noodles", "min_price": 1, "max_price": 3}
    ],
    "Cooking & Baking Supplies": [
        {"type": "Flour", "min_price": 2, "max_price": 6},
        {"type": "Sugar", "min_price": 2, "max_price": 5},
        {"type": "Spices & Seasonings", "min_price": 1, "max_price": 8},
        {"type": "Baking Powder", "min_price": 1, "max_price": 3},
        {"type": "Baking Soda", "min_price": 1, "max_price": 3},
        {"type": "Cooking Oils", "min_price": 3, "max_price": 10}
    ],
    "Dairy": [
        {"type": "Milk", "min_price": 2, "max_price": 4},
        {"type": "Cheese", "min_price": 3, "max_price": 10},
        {"type": "Yogurt", "min_price": 1, "max_price": 6},
        {"type": "Butter", "min_price": 2, "max_price": 6},
        {"type": "Cream", "min_price": 2, "max_price": 5},
        {"type": "Eggs", "min_price": 2, "max_price": 5}
    ],
    "Frozen Foods": [
        {"type": "Frozen Mixed Vegetables", "min_price": 2, "max_price": 5},
        {"type": "Frozen Mixed Fruits", "min_price": 3, "max_price": 7},
        {"type": "Vanilla Ice Cream", "min_price": 3, "max_price": 7},
        {"type": "Chocolate Ice Cream", "min_price": 3, "max_price": 7},
        {"type": "Strawberry Ice Cream", "min_price": 3, "max_price": 7},
        {"type": "Mint Chocolate Chip Ice Cream", "min_price": 4, "max_price": 8},
        {"type": "Cookies and Cream Ice Cream", "min_price": 4, "max_price": 8},
        {"type": "Frozen Burritos", "min_price": 3, "max_price": 7},
        {"type": "Frozen Chicken Tenders", "min_price": 5, "max_price": 10},
        {"type": "Frozen Macaroni and Cheese", "min_price": 4, "max_price": 8},
        {"type": "Frozen Meatballs", "min_price": 5, "max_price": 10},
        {"type": "Frozen Enchiladas", "min_price": 5, "max_price": 10},
        {"type": "Frozen Stir-Fry", "min_price": 5, "max_price": 10},
        {"type": "Frozen Pizza", "min_price": 4, "max_price": 10},
        {"type": "Frozen Mixed Seafood", "min_price": 5, "max_price": 15}
    ],
    "Meat & Seafood": [
        {"type": "Chicken Breast", "min_price": 5, "max_price": 10},
        {"type": "Chicken Thighs", "min_price": 4, "max_price": 9},
        {"type": "Whole Chicken", "min_price": 7, "max_price": 15},
        {"type": "Chicken Wings", "min_price": 4, "max_price": 8},
        {"type": "Ground Beef", "min_price": 6, "max_price": 12},
        {"type": "Beef Steak", "min_price": 10, "max_price": 25},
        {"type": "Beef Ribs", "min_price": 8, "max_price": 20},
        {"type": "Beef Brisket", "min_price": 8, "max_price": 18},
        {"type": "Pork Chops", "min_price": 5, "max_price": 12},
        {"type": "Pork Tenderloin", "min_price": 6, "max_price": 15},
        {"type": "Pork Shoulder", "min_price": 4, "max_price": 10},
        {"type": "Bacon", "min_price": 5, "max_price": 12},
        {"type": "Salmon", "min_price": 10, "max_price": 20},
        {"type": "Tuna", "min_price": 8, "max_price": 18},
        {"type": "Cod", "min_price": 7, "max_price": 15},
        {"type": "Haddock", "min_price": 7, "max_price": 14},
        {"type": "Trout", "min_price": 9, "max_price": 18},
        {"type": "Tilapia", "min_price": 5, "max_price": 12},
        {"type": "Swordfish", "min_price": 15, "max_price": 25},
        {"type": "Shrimp", "min_price": 8, "max_price": 20},
        {"type": "Sausage", "min_price": 4, "max_price": 10}
    ],
   "Produce": [
        {"type": "Apples", "min_price": 2, "max_price": 5},
        {"type": "Bananas", "min_price": 1, "max_price": 3},
        {"type": "Oranges", "min_price": 2, "max_price": 4},
        {"type": "Grapes", "min_price": 3, "max_price": 6},
        {"type": "Strawberries", "min_price": 3, "max_price": 7},
        {"type": "Blueberries", "min_price": 4, "max_price": 8},
        {"type": "Pineapples", "min_price": 3, "max_price": 8},
        {"type": "Mangoes", "min_price": 2, "max_price": 5},
        {"type": "Pears", "min_price": 2, "max_price": 4},
        {"type": "Peaches", "min_price": 2, "max_price": 5},
        {"type": "Carrots", "min_price": 1, "max_price": 3},
        {"type": "Broccoli", "min_price": 2, "max_price": 5},
        {"type": "Bell Peppers", "min_price": 2, "max_price": 4},
        {"type": "Lettuce", "min_price": 1, "max_price": 3},
        {"type": "Tomatoes", "min_price": 2, "max_price": 5},
        {"type": "Cucumbers", "min_price": 1, "max_price": 3},
        {"type": "Potatoes", "min_price": 1, "max_price": 4},
        {"type": "Onions", "min_price": 1, "max_price": 3},
        {"type": "Garlic", "min_price": 1, "max_price": 3},
        {"type": "Spinach", "min_price": 2, "max_price": 5}
    ],

    "Bakery": [
        {"type": "Whole Wheat Bread", "min_price": 2, "max_price": 5},
        {"type": "Sourdough Bread", "min_price": 3, "max_price": 6},
        {"type": "Rye Bread", "min_price": 3, "max_price": 6},
        {"type": "Croissants", "min_price": 2, "max_price": 4},
        {"type": "Danish Pastries", "min_price": 3, "max_price": 6},
        {"type": "Eclairs", "min_price": 3, "max_price": 5},
        {"type": "Cinnamon Rolls", "min_price": 3, "max_price": 5},
        {"type": "Plain Bagels", "min_price": 2, "max_price": 4},
        {"type": "Sesame Bagels", "min_price": 2, "max_price": 4},
        {"type": "Everything Bagels", "min_price": 2, "max_price": 4},
        {"type": "Dinner Rolls", "min_price": 2, "max_price": 5},
        {"type": "Burger Buns", "min_price": 2, "max_price": 5},
        {"type": "Hot Dog Buns", "min_price": 2, "max_price": 5},
        {"type": "Chocolate Cake", "min_price": 15, "max_price": 30},
        {"type": "Vanilla Cake", "min_price": 15, "max_price": 30},
        {"type": "Red Velvet Cake", "min_price": 20, "max_price": 35},
        {"type": "Cheesecake", "min_price": 20, "max_price": 35},
        {"type": "Apple Pie", "min_price": 12, "max_price": 25},
        {"type": "Cherry Pie", "min_price": 12, "max_price": 25},
        {"type": "Pumpkin Pie", "min_price": 10, "max_price": 20},
    ]
}


def generate_product_name(brand, product_type):
    """Generate a product name in the format 'Brand + Product Type'."""
    return f"{brand} {product_type}"

def generate_products(n=1000):
    conn = get_connection()
    cursor = conn.cursor()
    
    # Fetch category_ids and names from the categories table
    cursor.execute("SELECT category_id, category_name FROM categories")
    categories = cursor.fetchall()
    
    products = []
    
    for _ in range(n):
        category_id, category_name = random.choice(categories)
        product_detail = random.choice(detailed_grocery_products[category_name])
        buyin_price = round(random.uniform(product_detail['min_price'], product_detail['max_price']), 2)
        sell_price = round(buyin_price * random.uniform(1.1, 1.5), 2)
        brand = fake.company()  # Generate a random brand name
        product_name = generate_product_name(brand, product_detail['type'])
        
        # Append product data to the products list
        products.append((product_name, buyin_price, sell_price, category_id))
        
        # Insert product data into the database
        cursor.execute(
            "INSERT INTO products (product_name, buyin_price, sell_price, category_id) VALUES (%s, %s, %s, %s)",
            (product_name, buyin_price, sell_price, category_id)
        )
    
    conn.commit()
    cursor.close()
    conn.close()
    
    # Write the generated product data to a CSV file
    with open('C:/Users/Yihua/Desktop/5310 project/products.csv', 'w', newline='') as csvfile:
        csv_writer = csv.writer(csvfile)
        csv_writer.writerow(['product_name', 'buyin_price', 'sell_price', 'category_id'])
        for product in products:
            csv_writer.writerow(product)

if __name__ == "__main__":
    generate_products()
```
